Será que os dados acima influenciaram no interesse geral dos atletas pelo esporte ou realmente impactaram no desempenho deles? Podemos tentar descobrir se há algum tipo de correlação.

Você ainda possui o dataframe original contendo todos os atletas brasileiros, incluindo os sem medalha? Obtenha os valores médios de peso e altura por esporte daquele dataframe e compare-o com os dos medalhistas. Há alguma diferença significativa em algum esporte?

````python
df_historico_olimpiadas_br.groupby(by=['Sport'])['Height','Weight'].mean()
````

<font color='blue'>Remover o esporte que não tem média de peso nem média de altura</font>

````python
media_todos_atletas = df_historico_olimpiadas_br.groupby(by=['Sport'])['Height','Weight'].mean().drop('Art Competitions', axis=0)
````

**<font color='blue'>Como a lista de esportes na qual o Brasil já participou é maior do que a lista de esportes com medalhistas, é necessario filtrar apenas os esportes que já obtiveram medalhistas para fazer a comparação**</font>

````python
# Criação do filtro dos esportes com medalhas
lista_esportes_medalha = df_historico_olimpiadas_br[~ pd.isna(df_historico_olimpiadas_br['Medal'])].groupby(by=['Sport'])['Weight'].mean().index
filtro_esportes_medalha = media_todos_atletas.index.isin(lista_esportes_medalha)
````

````python
# Diferença entre média dos medalhistas brasileiros e a média de todos os atletas brasileiros por esporte (Altura e Peso)
df_historico_olimpiadas_br[~ pd.isna(df_historico_olimpiadas_br['Medal'])].groupby(by=['Sport'])['Height','Weight'].mean() - media_todos_atletas[filtro_esportes_medalha]
````

<font color='blue'>Para análise da tabela acima foi estabelecido o critério arbritrário para peso e altura:   
- A medida impacta no desempenho do atleta caso o módulo da diferença entre a média dos medalhistas e média de todos os atletas for maior ou igual a 7;
- Se o módulo da diferença é menor que 7, significa que a medida não tem relação com o desempenho do atleta</font>

#### <font color='blue'>Diferença de altura</font>
<font color='blue'>A partir do resultado acima, observa-se que a Natação e o Taekwondo são esportes onde a altura média dos medalhistas é significativamente maior que a altura média de todos os atletas olímpicos.</font>

#### <font color='blue'>Diferença de peso</font>
<font color='blue'>A partir do resultado acima, observa-se que o Taekwondo, a Ginástica, o Tiro ao alvo e o Basquete são esportes onde o peso médio dos atletas pode influenciar no desempenho.
- No Taekwondo e na Ginástica, os medalhistas são razoavelmente mais pesados que a média de peso dos atletas em geral;
- No Tiro e no Basquete, os medalhistas são razoavelmente mais leves que a média de peso dos atletas em geral;</font>

Existe um detalhe importante passando batido até agora em nossa análise: as categorias esportivas costumam ser divididas por gênero justamente por conta de diferenças físicas entre homens e mulheres que poderiam influenciar no desempenho. Compare a altura e peso médios de atletas brasileiros por esporte segmentado por sexo.

````python
media_todos_atletas_sexo = df_historico_olimpiadas_br.groupby(by=['Sport','Sex'])['Height','Weight'].mean().pivot_table(index="Sport", columns='Sex', values=['Height','Weight'])
media_medalhistas_sexo =  df_historico_olimpiadas_br[~ pd.isna(df_historico_olimpiadas_br['Medal'])].groupby(by=['Sport','Sex'])['Height','Weight'].mean().pivot_table(index="Sport", columns='Sex', values=['Height','Weight'])
media_medalhistas_sexo - media_todos_atletas_sexo[filtro_esportes_medalha]
````

<font color='blue'>Ao observar a tabela acima percebe-se que o módulo da diferença entre média dos medalhistas e média dos atletas em geral é significativamente maior para atletas do sexo masculino na maioria dos esportes. Ou seja, parece haver maior correlação entre a medida (peso e altura) e o desempenho para atletas do sexo masculino. Para atletas do sexo feminino parece haver pouca correlação entre desempenho e medidas (peso e altura) para a maioria dos esportes.</font>

## 3. Brasil vs Mundo

````python
pd.options.display.float_format = "{:,.2f}".format
````

Para finalizar, vamos fazer algumas comparações entre Brasil e mundo. Qual o ranking do Brasil em cada edição das olimpíadas? Lembrando que o ranking é ordenado por medalhas de ouro.

<font color='blue'>O index 'Rank' na tabela abaixo mostra a posição do Brasil em cada edição

````python
medalhas_paises_ano = pd.DataFrame(um_medalhista_p_evento_geral.groupby(by=['Year','NOC'])['Medal'].value_counts())
x = medalhas_paises_ano.rename(columns={"Medal": "Quantidade"})
x = x.pivot_table(index=['Year','NOC'],columns='Medal',values='Quantidade')
x = x[['Gold','Silver','Bronze']]
x[pd.isna(x)] = 0
x = x.sort_values(by=['Year','Gold','Silver','Bronze'], ascending=False, na_position='last')
x['Rank'] = x.groupby(by='Year').cumcount() + 1
x = x.set_index('Rank', append=True)
x[np.in1d(x.index.get_level_values(1), ['BRA'])].sort_values(by='Year')
````

Compare o maior medalhista em ouros do Brasil com o maior medalhista em ouros do mundo.

<font color='blue'>Essa e todas as outras questões de comparação onde o 'maior' é uma lista de pessoas, ou lista de esportes, a comparação foi feita utilizando a média dessas listas. Nesse primeiro caso, os maiores medalhistas brasileiros são uma lista de 13 pessoas. A média dessa lista será comparada com o maior medalhista mundial

**<font color='blue'>Maiores medalhistas ouro brasileiros**

````python
#dataframe filtrado apenas com os maiores medalhistas ouro
mais_medalhas_ouro_brasil = df_historico_olimpiadas_verao[df_historico_olimpiadas_verao['Name'].isin(mais_ouro_brasil.index)]

#dataframe com a quantidade de medalhas de cada um dos maiores medalhistas ouro
qtde_medalhas_MMOB = pd.DataFrame(mais_medalhas_ouro_brasil.groupby(by=['Name','Medal']).Medal.count()).rename(columns={'Medal':'Qtdade'})
qtde_medalhas_MMOB = qtde_medalhas_MMOB.pivot_table(index=['Name'], columns='Medal', values='Qtdade')
qtde_medalhas_MMOB['Silver'][pd.isna(qtde_medalhas_MMOB['Silver'])] = 0
qtde_medalhas_MMOB['Bronze'][pd.isna(qtde_medalhas_MMOB['Bronze'])] = 0
qtde_medalhas_MMOB['Total'] = qtde_medalhas_MMOB['Gold']+qtde_medalhas_MMOB['Silver']+qtde_medalhas_MMOB['Bronze']
qtde_medalhas_MMOB = qtde_medalhas_MMOB[['Gold','Silver','Bronze','Total']]

#dataframe com outras informacoes relevantes
outras_infos = mais_medalhas_ouro_brasil.groupby(by=['Name']).agg({'Sex':pd.Series.mode,'Age':'mean','Height':'mean','Weight':'mean','Event':'count'})
outras_infos.rename(columns={'Age':'Idade média nas edições','Event':'Participações'}, inplace=True)

#concatenar quantidade de medalhas com outras infos
maiores_medalhistas_ouro_brasil = pd.concat([qtde_medalhas_MMOB,outras_infos], axis=1)
maiores_medalhistas_ouro_brasil['Medals/Participações'] = maiores_medalhistas_ouro_brasil['Total']/maiores_medalhistas_ouro_brasil['Participações']*100
display1 = maiores_medalhistas_ouro_brasil.copy()
display1[['Gold','Silver','Bronze','Total','Height','Weight']] = display1[['Gold','Silver','Bronze','Total','Height','Weight']].applymap('{:,.0f}'.format)
display1['Idade média nas edições'] = display1['Idade média nas edições'].apply('{:,.2f}'.format)
display1['Medals/Participações'] = display1['Medals/Participações'].apply('{:,.2f}%'.format)
display1.style.set_properties(**{'text-align': 'center'})
````

**<font color='blue'>Maior medalhista ouro mundial**

````python
#mesmo processo para o maior medalhista ouro do mundo

mais_medalhas_ouro_mundo = df_historico_olimpiadas_verao[df_historico_olimpiadas_verao['Name'].isin(mais_ouro_mundo.index)]


#dataframe com a quantidade de medalhas do maior medalhista ouro
qtde_medalhas_MMOM = pd.DataFrame(mais_medalhas_ouro_mundo.groupby(by=['Name','Medal']).Medal.count()).rename(columns={'Medal':'Qtdade'})
qtde_medalhas_MMOM = qtde_medalhas_MMOM.pivot_table(index=['Name'], columns='Medal', values='Qtdade')
qtde_medalhas_MMOM['Silver'][pd.isna(qtde_medalhas_MMOM['Silver'])] = 0
qtde_medalhas_MMOM['Bronze'][pd.isna(qtde_medalhas_MMOM['Bronze'])] = 0
qtde_medalhas_MMOM['Total'] = qtde_medalhas_MMOM['Gold']+qtde_medalhas_MMOM['Silver']+qtde_medalhas_MMOM['Bronze']
qtde_medalhas_MMOM = qtde_medalhas_MMOM[['Gold','Silver','Bronze','Total']]


#dataframe com outras informacoes relevantes
outras_infos_MP = mais_medalhas_ouro_mundo.groupby(by=['Name']).agg({'Sex':pd.Series.mode,'Age':'mean','Height':'mean','Weight':'mean','Event':'count'})
outras_infos_MP.rename(columns={'Age':'Idade média nas edições','Event':'Participações'}, inplace=True)

#concatenar quantidade de medalhas com outras infos
maior_medalhista_ouro_mundo = pd.concat([qtde_medalhas_MMOM,outras_infos_MP], axis=1)
maior_medalhista_ouro_mundo['Medals/Participações'] = maior_medalhista_ouro_mundo['Total']/maior_medalhista_ouro_mundo['Participações']*100
display2 = maior_medalhista_ouro_mundo.copy()
display2[['Gold','Silver','Bronze','Total','Height','Weight']] = display2[['Gold','Silver','Bronze','Total','Height','Weight']].applymap('{:,.0f}'.format)
display2['Idade média nas edições'] = display2['Idade média nas edições'].apply('{:,.2f}'.format)
display2['Medals/Participações'] = display2['Medals/Participações'].apply('{:,.2f}%'.format)
display2.style.set_properties(**{'text-align': 'center'})
````

**<font color='blue'>Comparação entre os maiores medalhistas ouro brasileiros e o maior medalhista ouro mundial**  
<font color='blue'> A comparação é feita a partir das médias das caracteristicas dos medalhistas ouro brasileiros com o medalhista ouro mundial.


- Idade: é feito a média de idade do atleta quando este participou de mais de uma edição e em seguida é feito a média de idade entre os atletas;
- Participações: média da quantidade de eventos olimpícos em que os atletas brasileiros participaram


````python
comparacoes = ['Gold','Silver','Bronze','Total','Idade média nas edições','Height','Weight','Participações']
colunas = [comparacoes, ['Mundo','Brasil', 'Diferença']]
cols = pd.MultiIndex.from_product(colunas)

comparacao = pd.DataFrame(columns=cols)
for i in comparacoes:   
    comparacao[i,'Brasil'] = maiores_medalhistas_ouro_brasil[i]
    comparacao[i,'Mundo'] = maiores_medalhistas_ouro_brasil[i].apply(lambda x:maior_medalhista_ouro_mundo[i])
    comparacao[i,'Diferença'] = maiores_medalhistas_ouro_brasil[i].apply(lambda x:maior_medalhista_ouro_mundo[i]-x)
    
comparacao.loc['Média'] = comparacao.mean()
comparacao = pd.DataFrame(comparacao.loc['Média'])
comparacao.index.names = ['Características','Comparação']
comparacao = comparacao.pivot_table(index='Características', columns='Comparação', values='Média')[['Mundo','Brasil','Diferença']]

#Média ponderada pela quantidade de participações da porcentagem de medalhas em relação à quantidade de participações
comparacao.loc['Medals/Participações'] = comparacao.loc['Total']/comparacao.loc['Participações']*100
comparacao['Diferença'].loc['Medals/Participações'] = comparacao['Mundo'].loc['Medals/Participações'] - comparacao['Brasil'].loc['Medals/Participações']

#formatação
comparacao.loc['Medals/Participações'] = comparacao.loc['Medals/Participações'].apply('{:,.2f}%'.format)
comparacao.loc[['Gold','Silver','Bronze','Total','Participações','Medals/Participações','Height','Weight','Idade média nas edições']]
````

````python
comparacoes = ['Gold', 'Silver', 'Bronze', 'Total', 'Idade média nas edições', 'Height', 'Weight', 'Participações']
colunas_fem = [comparacoes, ['Mundo', 'Feminino', 'Dif Fem']]
colunas_masc = [comparacoes, ['Masculino', 'Dif Masc']]

cols_fem = pd.MultiIndex.from_product(colunas_fem)
cols_masc = pd.MultiIndex.from_product(colunas_masc)
                
maiores_medalhistas_ouro_brasil_fem = maiores_medalhistas_ouro_brasil[maiores_medalhistas_ouro_brasil['Sex'] == 'F']
maiores_medalhistas_ouro_brasil_masc = maiores_medalhistas_ouro_brasil[maiores_medalhistas_ouro_brasil['Sex'] == 'M']

comparacao_fem = pd.DataFrame(columns=cols_fem)
comparacao_masc = pd.DataFrame(columns=cols_masc)

for i in comparacoes:
    comparacao_fem[i, 'Feminino']= maiores_medalhistas_ouro_brasil_fem[i]
    comparacao_fem[i, 'Mundo']= maiores_medalhistas_ouro_brasil_fem[i].apply(lambda x:maior_medalhista_ouro_mundo[i])
    comparacao_fem[i, 'Dif Fem']=maiores_medalhistas_ouro_brasil_fem[i].apply(lambda x:maior_medalhista_ouro_mundo[i]-x)

    comparacao_masc[i, 'Masculino']=maiores_medalhistas_ouro_brasil_masc[i]
    comparacao_masc[i, 'Dif Masc']=maiores_medalhistas_ouro_brasil_masc[i].apply(lambda x:maior_medalhista_ouro_mundo[i]-x)

comparacao_fem = pd.DataFrame(comparacao_fem.mean())
comparacao_fem.rename(columns={0:'valores'},inplace=True)
comparacao_fem.index.names = ['Medidas','Média']
comparacao_fem = comparacao_fem.pivot_table(index='Medidas', columns='Média', values='valores')[['Mundo','Feminino','Dif Fem']]

comparacao_masc = pd.DataFrame(comparacao_masc.mean())
comparacao_masc.rename(columns={0:'valores'},inplace=True)
comparacao_masc.index.names = ['Medidas','Média']
comparacao_masc = comparacao_masc.pivot_table(index='Medidas', columns='Média', values='valores')[['Masculino','Dif Masc']]

# concatenar
comparacao_sexo = pd.concat([comparacao_fem, comparacao_masc], axis=1)

# Média ponderada pela quantidade de participações da porcentagem de medalhas em relação à quantidade de participações
comparacao_sexo.loc['Medals/Participações'] = comparacao_sexo.loc['Total']/comparacao_sexo.loc['Participações']*100
comparacao_sexo['Dif Fem'].loc['Medals/Participações'] = comparacao_sexo['Mundo'].loc['Medals/Participações'] - comparacao_sexo['Feminino'].loc['Medals/Participações']
comparacao_sexo['Dif Masc'].loc['Medals/Participações'] = comparacao_sexo['Mundo'].loc['Medals/Participações'] - comparacao_sexo['Masculino'].loc['Medals/Participações']

# formatação
comparacao_sexo.loc['Medals/Participações'] = comparacao_sexo.loc['Medals/Participações'].apply('{:,.2f}%'.format)
comparacao_sexo = comparacao_sexo.loc[['Gold','Silver','Bronze','Total','Participações','Medals/Participações','Height','Weight','Idade média nas edições']]
comparacao_sexo = comparacao_sexo[['Mundo','Feminino','Masculino','Dif Fem','Dif Masc']]

comparacao_sexo
````

````python
mais_medalhas_mundo_esporte_mais_medalhas_br = df_historico_olimpiadas_verao[df_historico_olimpiadas_verao['Name']==maiores_mundo_mais_medalhas_br.index[0]]

# #dataframe com a quantidade de medalhas de cada um dos maiores medalhistas ouro
qtde_medalhas_MMMEM = pd.DataFrame(mais_medalhas_mundo_esporte_mais_medalhas_br.groupby(by=['Name','Medal']).Medal.count()).rename(columns={'Medal':'Qtdade'})
qtde_medalhas_MMMEM = qtde_medalhas_MMMEM.pivot_table(index=['Name'], columns='Medal', values='Qtdade')
qtde_medalhas_MMMEM['Bronze'] = 0
qtde_medalhas_MMMEM['Total'] = qtde_medalhas_MMMEM['Gold'] + qtde_medalhas_MMMEM['Silver'] + qtde_medalhas_MMMEM['Bronze']
qtde_medalhas_MMMEM = qtde_medalhas_MMMEM[['Gold','Silver','Bronze','Total']]
qtde_medalhas_MMMEM

#dataframe com outras informacoes relevantes
outras_infos = mais_medalhas_mundo_esporte_mais_medalhas_br.groupby(by=['Name']).agg({'Sex':pd.Series.mode,'Age':'mean','Height':'mean','Weight':'mean','Event':'count'})
outras_infos.rename(columns={'Age':'Idade média nas edições','Event':'Participações'}, inplace=True)

#concatenar quantidade de medalhas com outras infos
maiores_medalhistas_mundo_esporte_mais_medalhas_br = pd.concat([qtde_medalhas_MMMEM,outras_infos], axis=1)
maiores_medalhistas_mundo_esporte_mais_medalhas_br['Medals/Participações'] = maiores_medalhistas_mundo_esporte_mais_medalhas_br['Total']/maiores_medalhistas_mundo_esporte_mais_medalhas_br['Participações']*100
#maiores_medalhistas_mundo_esporte_mais_medalhas_br.loc['Média'] = maiores_medalhistas_mundo_esporte_mais_medalhas_br.mean()
display6 = maiores_medalhistas_mundo_esporte_mais_medalhas_br.copy()
display6[['Gold','Silver','Bronze','Total','Height','Weight']] = display6[['Gold','Silver','Bronze','Total','Height','Weight']].applymap('{:,.0f}'.format)
display6['Idade média nas edições'] = display6['Idade média nas edições'].apply('{:,.2f}'.format)
display6['Medals/Participações'] = display6['Medals/Participações'].apply('{:,.2f}%'.format)
display6.style.set_properties(**{'text-align': 'center'})
````

````python
comparacoes = ['Gold','Silver','Bronze','Total','Idade média nas edições','Height','Weight','Participações']
colunas = [comparacoes, ['Mundo','Brasil', 'Diferença']]
cols = pd.MultiIndex.from_product(colunas)
comparacao = pd.DataFrame(columns=cols)
for i in comparacoes:   
    comparacao[i,'Brasil'] = maiores_medalhistas_br_esporte_mais_medalhas_br[i]
    comparacao[i,'Mundo'] = maiores_medalhistas_br_esporte_mais_medalhas_br[i].apply(lambda x:maiores_medalhistas_mundo_esporte_mais_medalhas_br[i])
    comparacao[i,'Diferença'] = maiores_medalhistas_br_esporte_mais_medalhas_br[i].apply(lambda x:maiores_medalhistas_mundo_esporte_mais_medalhas_br[i]-x)

comparacao    
comparacao.loc['Média'] = comparacao.mean()
comparacao = pd.DataFrame(comparacao.loc['Média'])

comparacao.index.names = ['Características','Comparação']
comparacao = comparacao.pivot_table(index='Características', columns='Comparação', values='Média')[['Mundo','Brasil','Diferença']]
comparacao

# #Média ponderada pela quantidade de participações da porcentagem de medalhas em relação à quantidade de participações
comparacao.loc['Medals/Participações'] = comparacao.loc['Total']/comparacao.loc['Participações']*100
comparacao['Diferença'].loc['Medals/Participações'] = comparacao['Mundo'].loc['Medals/Participações'] - comparacao['Brasil'].loc['Medals/Participações']

# #formatação
comparacao.loc['Medals/Participações'] = comparacao.loc['Medals/Participações'].apply('{:,.2f}%'.format)
comparacao.loc[['Gold','Silver','Total','Participações','Medals/Participações','Height','Weight','Idade média nas edições']]
````
